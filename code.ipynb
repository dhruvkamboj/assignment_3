```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv',encoding="latin1")
data.head()
```

```python
x=data['no2'].dropna()
```

```python
r=102303645
ar=0.05*(r%7)
br=0.3*(r%5+1)

z=x+ar*np.sin(br*x)
```

```python
variance=np.var(z)

muvalue=np.mean(z)
cvalue=1/np.sqrt(2*np.pi*variance)
lambdavalue=1/(2*variance)
```

```python
print(lambdavalue)
print(muvalue)
print(cvalue)
```
